pandas digunakan untuk membaca dan mengolah dataset.
numpy digunakan untuk operasi angka.
train_test_split digunakan untuk membagi data menjadi data latih dan data validasi.
accuracy_score, classification_report, dan confusion_matrix digunakan untuk mengevaluasi hasil model.

Selain itu, digunakan juga tiga model klasifikasi, yaitu Logistic Regression, KNN, dan Decision Tree.

3. Membaca Dataset

```python
train = pd.read_csv("data_training.csv")
test = pd.read_csv("data_testing.csv")

display(train.head())
display(test.head())

print("Ukuran data training:", train.shape)
print("Ukuran data testing:", test.shape)
```

Pada langkah ini, file CSV dibaca menggunakan pd.read_csv().

train.head() digunakan untuk menampilkan 5 data pertama dari data training.
test.head() digunakan untuk menampilkan 5 data pertama dari data testing.

train.shape dan test.shape digunakan untuk melihat jumlah baris dan kolom pada masing-masing dataset.

4. Mengecek Informasi Dataset

```python
print("Informasi data training:")
train.info()

print("\nInformasi data testing:")
test.info()
```

Langkah ini dilakukan untuk melihat informasi umum dataset, seperti nama kolom, jumlah data, dan tipe data.

Dari tahap ini, kita dapat mengetahui apakah semua fitur berbentuk numerik atau ada kolom yang perlu diperbaiki terlebih dahulu.

5. Mengecek Missing Value

```python
print("Missing value data training:")
display(train.isnull().sum())

print("Missing value data testing:")
display(test.isnull().sum())
```

Missing value adalah data yang kosong atau tidak memiliki nilai.

Pengecekan ini penting karena model machine learning tidak bisa bekerja dengan baik jika masih terdapat data kosong. Jika ada missing value, maka akan ditangani menggunakan SimpleImputer pada pipeline model.

Pada kode ini, missing value dicek menggunakan:

```python
train.isnull().sum()
```

Artinya, program menghitung jumlah nilai kosong pada setiap kolom.

6. Membersihkan Nama Kolom

```python
train.columns = train.columns.str.strip()
test.columns = test.columns.str.strip()
```

Langkah ini digunakan untuk menghapus spasi yang tidak sengaja ada pada nama kolom.

Contohnya, jika ada kolom bernama " quality " maka akan diubah menjadi "quality".

Hal ini penting agar tidak terjadi error saat memanggil nama kolom.

7. Menentukan Kolom ID dan Target

```python
id_col = None

for col in test.columns:
    if col.lower() == "id":
        id_col = col
        break

target_col = "quality"
```

Pada data testing biasanya terdapat kolom id. Kolom ini digunakan sebagai identitas setiap data.

Kolom quality digunakan sebagai target karena nilai inilah yang ingin diprediksi.

```python
# id = identitas data
# quality = target yang diprediksi
```

8. Memisahkan Fitur dan Target

```python
drop_cols = [target_col]

for col in train.columns:
    if col.lower() == "id":
        drop_cols.append(col)

X = train.drop(columns=drop_cols)
Y = train[target_col]

X_test = test.drop(columns=[id_col])

print("Fitur yang digunakan:")
print(X.columns)

print("\nDistribusi target:")
display(Y.value_counts().sort_index())
```

data dipisahkan menjadi fitur dan target.

X berisi fitur-fitur kimiawi wine yang digunakan untuk memprediksi kualitas wine.
Y berisi target, yaitu kolom quality.

Kolom id tidak digunakan sebagai fitur karena hanya sebagai identitas data, bukan faktor yang memengaruhi kualitas wine.

Data testing juga dipisahkan dengan menghapus kolom id, sehingga hanya menyisakan fitur yang akan dipakai untuk prediksi.

9. Membagi Data Training dan Validasi

```python
X_train, X_val, Y_train, Y_val = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42,
    stratify=Y
)
```

Data training dibagi menjadi dua bagian:

```python
# 80% data latih
# 20% data validasi
```

Data latih digunakan untuk melatih model.
Data validasi digunakan untuk menguji model sementara.

Parameter random_state=42 digunakan agar hasil pembagian data tetap sama setiap kali kode dijalankan.

Parameter stratify=Y digunakan agar proporsi setiap kelas quality tetap seimbang antara data latih dan data validasi.

10. Membuat Model Logistic Regression

```python
model_logreg = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000))
])

model_logreg.fit(X_train, Y_train)
Y_pred_logreg = model_logreg.predict(X_val)

acc_logreg = accuracy_score(Y_val, Y_pred_logreg)
print("Accuracy Logistic Regression:", acc_logreg)
```

Logistic Regression adalah model klasifikasi yang digunakan untuk memprediksi kelas berdasarkan hubungan antar fitur.

Pada model ini digunakan pipeline yang berisi:

```python
# SimpleImputer  → menangani missing value
# StandardScaler → melakukan scaling data
# Model          → Logistic Regression
```

Scaling diperlukan karena Logistic Regression sensitif terhadap perbedaan skala antar fitur.

Setelah model dilatih menggunakan X_train dan Y_train, model memprediksi data validasi X_val. Hasil prediksi kemudian dibandingkan dengan Y_val menggunakan accuracy.

11. Membuat Model KNN

```python
model_knn = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", KNeighborsClassifier(n_neighbors=5))
])

model_knn.fit(X_train, Y_train)
Y_pred_knn = model_knn.predict(X_val)

acc_knn = accuracy_score(Y_val, Y_pred_knn)
print("Accuracy KNN:", acc_knn)
```

KNN atau K-Nearest Neighbors adalah model yang memprediksi kelas berdasarkan tetangga terdekat.

Pada kode ini digunakan n_neighbors=5, artinya model melihat 5 data terdekat untuk menentukan kelas prediksi.

KNN juga membutuhkan scaling karena model ini menghitung jarak antar data. Jika skala fitur berbeda jauh, hasil prediksi bisa kurang baik.

12. Membuat Model Decision Tree

```python
model_tree = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("model", DecisionTreeClassifier(random_state=42))
])

model_tree.fit(X_train, Y_train)
Y_pred_tree = model_tree.predict(X_val)

acc_tree = accuracy_score(Y_val, Y_pred_tree)
print("Accuracy Decision Tree:", acc_tree)
```

Decision Tree adalah model klasifikasi berbentuk pohon keputusan.

Model ini bekerja dengan membuat aturan berdasarkan fitur-fitur yang ada.
Contohnya, model akan memisahkan data berdasarkan nilai alkohol, acidity, sulphates, dan fitur lainnya.

Pada Decision Tree, scaling tidak wajib digunakan karena model tidak menghitung jarak seperti KNN.

13. Membandingkan Akurasi Model

```python
evaluasi_model = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "KNN",
        "Decision Tree"
    ],
    "Accuracy": [
        accuracy_score(Y_val, Y_pred_logreg),
        accuracy_score(Y_val, Y_pred_knn),
        accuracy_score(Y_val, Y_pred_tree)
    ]
})

evaluasi_model = evaluasi_model.sort_values(
    by="Accuracy",
    ascending=False
).reset_index(drop=True)

display(evaluasi_model)
```

Pada langkah ini, hasil akurasi dari ketiga model dimasukkan ke dalam tabel.

Tabel ini digunakan untuk melihat model mana yang memiliki akurasi paling tinggi.

Model dengan nilai accuracy tertinggi akan dipilih sebagai model terbaik.

13. Memilih Model Terbaik

```python
model_terbaik = evaluasi_model.loc[0, "Model"]

print("Model terbaik adalah:", model_terbaik)

if model_terbaik == "Logistic Regression":
    final_model = model_logreg
    Y_pred_best = Y_pred_logreg

elif model_terbaik == "KNN":
    final_model = model_knn
    Y_pred_best = Y_pred_knn

else:
    final_model = model_tree
    Y_pred_best = Y_pred_tree
```

Karena tabel evaluasi sudah diurutkan dari accuracy tertinggi ke terendah, maka baris pertama dianggap sebagai model terbaik.

Model terbaik ini nantinya digunakan untuk evaluasi lebih lanjut dan prediksi data testing.

15. Evaluasi Model Terbaik

```python
print("Classification Report:")
print(classification_report(Y_val, Y_pred_best))

print("Confusion Matrix:")
display(pd.DataFrame(confusion_matrix(Y_val, Y_pred_best)))
```

Pada tahap ini, model terbaik dievaluasi lebih detail.

classification_report menampilkan precision, recall, f1-score, dan accuracy.

confusion_matrix menunjukkan perbandingan antara nilai aktual dan nilai prediksi.

Dari hasil ini, kita bisa melihat apakah model sudah cukup baik dalam memprediksi setiap kelas quality.

16. Melatih Ulang Model Terbaik dengan Semua Data Training

```python
if model_terbaik == "Logistic Regression":
    final_model = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
        ("model", LogisticRegression(max_iter=1000))
    ])

elif model_terbaik == "KNN":
    final_model = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
        ("model", KNeighborsClassifier(n_neighbors=5))
    ])

else:
    final_model = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", DecisionTreeClassifier(random_state=42))
    ])

final_model.fit(X, Y)

print("Model final berhasil dilatih menggunakan seluruh data training.")
```

Setelah model terbaik dipilih, model tersebut dilatih ulang menggunakan seluruh data training.

Tujuannya supaya model belajar dari semua data yang tersedia, bukan hanya dari 80% data latih.

Dengan cara ini, model diharapkan menghasilkan prediksi yang lebih baik pada data testing.

17. Prediksi Data Testing

```python
# ============================================================
# PREDIKSI DATA TESTING
# ============================================================

prediksi_quality = final_model.predict(X_test)

hasil_prediksi = pd.DataFrame({
    "id": test[id_col],
    "quality": prediksi_quality
})

# Menampilkan semua data hasil prediksi
display(hasil_prediksi)
```
